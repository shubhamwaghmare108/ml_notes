```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Load Titanic dataset
df = sns.load_dataset('titanic')

# Select numeric useful features
df = df[['age','fare','pclass','sibsp','parch']]

# Handle missing values
df = df.dropna()
```

```python

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(df)
```

```python
# Compute linkage matrix
Z = linkage(X, method='ward')
```

```python
# Plot dendrogram
plt.figure(figsize=(12,6))
dendrogram(Z)

plt.title("Titanic Dataset Dendrogram")
plt.xlabel("Passengers")
plt.ylabel("Distance")
plt.show()
```

```python
plt.figure(figsize=(12,6))
dendrogram(Z)

plt.axhline(y=30, color='r', linestyle='--')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

clusters = fcluster(Z, t=10, criterion='distance')
df['cluster'] = clusters
```

```python
df.groupby('cluster').mean()
```

```python
Method 1  Max Distance Gap (Elbow on Dendrogram)

Look at the largest jump in merge distances in the linkage matrix.

Big jump = merging very different clusters  cut just before it.
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage

Z = linkage(X, method='ward')

# distances of merges
distances = Z[:,2]

# find largest gap
gaps = np.diff(distances)
cut_index = np.argmax(gaps)

cut_height = distances[cut_index]
print("Auto cut height:", cut_height)
```

```python
plt.figure(figsize=(12,6))
dendrogram(Z)

plt.axhline(y=cut_height, color='r', linestyle='--')
plt.show()
```

```python
Largest distance jump = clusters being forced together.
```

```python
Method 2  Silhouette Score (Best in practice )
Try different cluster counts and choose the one with highest silhouette score.
```

```python
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import fcluster

best_score = -1
best_k = 2
#"Because silhouette score ranges from -1 to 1, initializing with -1 ensures any valid score will replace it.Atleast two clusters are required for silhoutee"
for k in range(2,10):
    labels = fcluster(Z, k, criterion='maxclust')
    score = silhouette_score(X, labels)
    
    if score > best_score:
        best_score = score
        best_k = k

print("Best clusters:", best_k)
print("Best silhouette:", best_score)
```

```python
Method 3  Inconsistency Method (SciPy built-in)
Detects abnormal merges in dendrogram.
```

```python
from scipy.cluster.hierarchy import inconsistent

inc = inconsistent(Z)

cut_height = np.mean(inc[:,3])  # inconsistency values
print(cut_height)
```

```python
Method 4  Calinski-Harabasz Index
Higher = better clustering
```

```python
from sklearn.metrics import calinski_harabasz_score

best_score = 0
best_k = 2

for k in range(2,10):
    labels = fcluster(Z, k, criterion='maxclust')
    score = calinski_harabasz_score(X, labels)

    if score > best_score:
        best_score = score
        best_k = k

print(best_k)
```
