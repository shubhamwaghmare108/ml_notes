```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data from the CSV file
df = pd.read_csv("NIFTY 50_Historical_PR_10092024to10092025.csv")
```

```python
df
```

```python
# Create a new feature 'current_close' by shifting the 'Close' column
df["current_close"] = df["Close"].shift(1)
```

```python
df
```

```python
# Drop rows with NaN values and drop the unnecessary columns from X
df.at[0,"current_close"]=[25000]
X = df.drop(columns=["Date", "Index Name",  "current_close"])
y = df["current_close"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train the SGDRegressor model on the scaled data
model = SGDRegressor(max_iter=1000, eta0=0.01)
model.fit(X_train_scaled, y_train)
```

```python
# Print the learned parameters
print(f"Learned coefficients: {model.coef_}")
print(f"Learned intercept: {model.intercept_[0]:.4f}")
```

```python
model.predict([[24999.40,	25130.50,	24896.80,	25041.10]])
```

```python
# You can also evaluate the model's performance
print(f"\nModel score on test data: {model.score(X_test_scaled, y_test):.4f}")
```
