```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Sample data (same logic as earlier example but 2D)
X = np.array([
    [1, 2],
    [2, 3],
    [8, 7],
    [9, 8],
    [1, 0]
])
```

```python
# Step 1: Compute linkage matrix
Z = linkage(X, method='ward')  
# methods: 'single', 'complete', 'average', 'ward'
```

```python
# Step 2: Plot dendrogram
plt.figure(figsize=(8,5))
dendrogram(Z)

plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
Look for large vertical gap without horzontal cutin dendrogram.
Cut horizontally there
That gives optimal clusters
```

```python
plt.figure(figsize=(8,5))
dendrogram(Z)

plt.axhline(y=5, color='r', linestyle='--')  # cut line
plt.show()
```

```python
labels = ['A','B','C','D','E']

dendrogram(Z, labels=labels)
plt.show()
```
